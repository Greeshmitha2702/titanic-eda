```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = sns.load_dataset('titanic')
titanic_data
```

```python
print(titanic_data.shape)
print()
print(titanic_data.info())
```

```python
print(titanic_data.isnull().sum())
print()
print(titanic_data.describe())
```

```python
#To find no of males and no of females
sns.countplot(titanic_data,x='sex')
```

```python
#sex-wise survival counts
sns.countplot(titanic_data,x='sex',hue='survived')
```

```python
#class wise survival counts
sns.countplot(titanic_data,x='class',hue='survived')
```

```python
#to plot class wise mean age
sns.pointplot(titanic_data,y='class',x='age')
```

```python
#correlation map b/w class and fare
class_map = {'First':1,'Second':2,'Third':3}
titanic_data['class_num'] = titanic_data['class'].map(class_map)
new_tit = titanic_data[['class_num','fare']]
sns.heatmap(new_tit.corr(),annot=True)
```

```python
#to show age distribution
titanic_data['age'].hist()
#conclusion: Most of the passengers are b/w 20-40 age group
```

```python
#To findout survival based on age group
titanic_data.groupby('age')['survived'].mean()
#conclusion: Children had a higher survival rate compared to older passengers.
```

```python
# Based on embark_town, count of survived males and females
town_survival = (
    titanic_data[titanic_data['survived'] == 1]
    .groupby(['embark_town', 'sex'])
    .size()
    .unstack() #creates 2D table
    .rename(columns = {'male':'Male Survived','female':'Female Survived'})
)
town_survival
```

```python
#Based on family size, how many belongs to classes(eg: family size:1=> class 1:how many, class2:how many, class3:how many)
titanic_data['family_size'] = 1 + titanic_data['sibsp'] + titanic_data['parch']
family_survived = titanic_data.groupby(['family_size','pclass']).size().unstack(fill_value=0)
#fill_value helps to fill the NaN values with specified val
print(family_survived)
sns.heatmap(family_survived,fmt='d',annot=True,cmap='YlGnBu')
```

```python
#NUMPY
avg_age = np.mean(titanic_data['age']).round(2)
print("Average age of the passengers is",avg_age)
print()
fare = np.percentile(titanic_data['fare'].dropna(),25)
#25% of passengers paid fare ≤ 7.9
print(fare)
```

```python
#Survival rate by sex and class
sex_class_survival =(
        titanic_data[titanic_data['survived']==1]
        .groupby(['class','sex'])
        .size()
        .unstack()
)
sex_class_survival
```

```python
missing_count = titanic_data.isna().sum()
missing_percentage = (missing_count / len(titanic_data) * 100).round(2)
missing_table = pd.DataFrame({'Column':missing_percentage.index,'Missing count':missing_count.values,'Missing %':missing_percentage.values})
missing_table
```

```python
fig, axs = plt.subplots(2,1)
survived = titanic_data[titanic_data['survived']==1]['age']
non_survived = titanic_data[titanic_data['survived']==0]['age']
axs[0].hist(survived,color = 'green')
axs[1].hist(non_survived,color = 'red')
```

```python
#Boxplot comparison: fare by class
sns.boxplot(data = titanic_data, x = 'pclass',y = 'fare')
plt.title('Fare by Passenger Class')
```

```python
sns.countplot(titanic_data,x='embark_town',hue = 'survived')
plt.title('Survivor Count by Embark Town')
```

```python
numeric = titanic_data[['survived','pclass','age','fare','sibsp','parch','family_size']]
sns.heatmap(numeric.corr(),cmap='coolwarm',annot=True)
```
